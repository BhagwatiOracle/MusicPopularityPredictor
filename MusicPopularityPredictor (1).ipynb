```python
#Outlier removal
print('Highest allowed',df['Loudness']. mean()+3*df['Loudness'].std())
print("Lowest allowed",df['Loudness'].mean()-3*df['Loudness'].std())
```

```python
lowerlimit=df['Loudness'].mean()-3*df['Loudness'].std()
upperlimit=df['Loudness'].mean()+3*df['Loudness'].std()
```

```python
import numpy as np
df['Loudness']=np.where(df['Loudness']>upperlimit,
            upperlimit,
            np.where(df['Loudness']<lowerlimit,
                     lowerlimit,
                     df['Loudness']))
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,PowerTransformer
from sklearn.metrics import mean_squared_error,r2_score
```

```python
features=['Energy','Valence','Danceability','Loudness','Acousticness','Tempo','Speechiness','Liveness','Duration (ms)']
x=df[features]
y=df['Popularity']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)

```

```python
x_train.shape
```

```python
from sklearn.preprocessing import FunctionTransformer
def OutlierRemoval(X):
  features=['Energy','Valence','Danceability','Loudness','Acousticness','Tempo','Speechiness','Liveness','Duration (ms)']

  df=pd.DataFrame(X,columns=features)
  Q1=np.percentile(X,25)
  Q3=np.percentile(X,75)
  iqr=Q3-Q1
  upper_limit = Q3 + 1.5 * iqr
  lower_limit = Q1 - 1.5 * iqr

  df=np.where(df>upperlimit,upperlimit,np.where(X<lower_limit,lower_limit,X))
  return df
fc=FunctionTransformer(OutlierRemoval)
```

```python
x_train_trf=fc.fit_transform(x_train)
x_train_trf.shape
```

```python
#PowerTransformation & StandardScaling
from sklearn.compose import ColumnTransformer
features=['Energy','Valence','Danceability','Loudness','Tempo','Speechiness','Liveness','Duration (ms)']
tf=ColumnTransformer([
    ('trf1',PowerTransformer(),[4]),
    ('trf2',StandardScaler(),[0,1,2,3,5,6,7,8])
],remainder='passthrough')
x_train=tf.fit_transform(x_train)
x_test=tf.transform(x_test)
```

```python
x_train.shape
```

## Model traning

```python
#Model Trainning
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
param_grid={'alpha':[0.001,0.01,0.1,1,10,100]}
model=Ridge()
grid_search=GridSearchCV(model,param_grid,cv=5)
grid_search.fit(x_train,y_train)
y_pred=grid_search.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso
param_grid={'alpha':[0.001,0.01,0.1,1,10,100]}
model=Lasso()
grid_search=GridSearchCV(model,param_grid,cv=5)
grid_search.fit(x_train,y_train)
y_pred=grid_search.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best','random'],
    'max_depth':[2,4,8,10,None],
    'min_samples_split':[2,4,8,10],
    'min_samples_leaf':[1,2,3,4],
    'max_features':[None,'sqrt','log2']
    }
grid_search=GridSearchCV(DecisionTreeRegressor(),param_grid,cv=5)
grid_search.fit(x_train,y_train)
y_pred=grid_search.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import VotingRegressor
vr=VotingRegressor([
    ('lr',LinearRegression()),
    ('ridge',Ridge()),
    ('lasso',Lasso()),
    ('dt',DecisionTreeRegressor())
])
vr.fit(x_train,y_train)
y_pred=grid_search.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

param_grid={
    'n_estimators':[10,20,60,100,120],
    'max_features':[None,'sqrt','log2'],
    'max_depth':[2,4,8,10,None],
    'max_samples':[0.5,0.75,1.0],
    'min_samples_split':[2,4,8,10],
    'min_samples_leaf':[1,2,3,4]
}

model=RandomizedSearchCV(RandomForestRegressor(),param_grid,cv=5,n_jobs=-1)
model.fit(x_train,y_train)
y_pred=grid_search.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

```python
from xgboost import XGBRegressor
from scipy.stats import uniform,randint
from sklearn.model_selection import RandomizedSearchCV
param_grid={
    'max_depth':randint(3,10),
    'n_estimators':randint(50,150),
    'eta':uniform(0.01,0.3),
    'gamma':uniform(0,0.05),
    'subsample':uniform(0.6,0.4),
    'colsample_bytree':uniform(0.6,0.4),
    'tree_method':['hist'],
    'n_jobs':[-1]
}

cv=RandomizedSearchCV(XGBRegressor(),param_grid,cv=5)
cv.fit(x_train,y_train)
y_pred=cv.predict(x_test)
print('Mean Squared Error',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```
